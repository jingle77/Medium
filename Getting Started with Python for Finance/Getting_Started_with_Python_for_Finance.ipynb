```python
## Libraries
import yfinance as yf
import pandas as pd

## Data
aapl = yf.Ticker("AAPL")
aapl_df = aapl.history(start="2013-01-01", end="2023-01-01")
```

```python
aapl_df.info()
```

```python
aapl_df.head()
```

```python
aapl_daily_returns = aapl_df[['Close']].pct_change()
avg_annual_return = aapl_daily_returns.mean()[0] * 250
annual_variance = aapl_daily_returns.var() * 250
```

```python
avg_annual_return_formatted = "{:0.0%}".format(avg_annual_return)
print(avg_annual_return_formatted)
```
